```python
def generate_months(year):
    dates_ser  = pd.Series(pd.period_range("1/1/" + str(year+7), freq="M", periods=12))
    dates_ser = pd.to_datetime(dates_ser.astype(str) + '-01', format='%Y-%m-%d')   
    df = pd.DataFrame(dates_ser)
    df.rename(columns = {0:'date'}, inplace = True)
#    df.set_index('date')
    return dates_ser
    
```

```python
def generate_data_for_drink(year_consumption_one_type): 
    temp = pd.Series(index=range(12)) #пустой series
    dt = (calculate_monthly_consumption(year_consumption_one_type))
    temp = pd.Series(data = dt, index=temp.index) # в пустой series дописали помесячное потребление    
    return temp
    
```

```python
def generate_data_early_for_region(year,region,year_consumption):
    df = pd.DataFrame(index=range(12), columns=['dt','region','wine','beer','vodka','champagne','brandy'], dtype=float)
    df.dt = generate_months(year)
    df.region = region
    df.wine = generate_data_for_drink(year_consumption[0])
    df.beer = generate_data_for_drink(year_consumption[1])
    df.vodka = generate_data_for_drink(year_consumption[2])
    df.champagne = generate_data_for_drink(year_consumption[3])
    df.brandy = generate_data_for_drink(year_consumption[4])
    return df
```

```python
def generate_data_for_region(region):
    df = df_wo_nan[df_wo_nan['region'] == region]
    result = pd.concat([ generate_data_early_for_region(row.year, row.region, [row.wine,row.beer,row.vodka,row.champagne,row.brandy]) for index, row in df.iterrows()], ignore_index=True)
       
    return result
```

```python
def generate_data():
    regions = df_wo_nan['region'].unique()
    result = pd.concat([generate_data_for_region(region) for region in regions],ignore_index=True)
    return result
```
